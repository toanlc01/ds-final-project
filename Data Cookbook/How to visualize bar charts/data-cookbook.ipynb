# How to visualize a bar charts

In the other rescipe, we have learned too add a `level` column based on the experience. I will copy the code below and if you want more detailed explanation, please visit the other rescipe

```python
# Import the libraries
import pandas as pd

# Read the data into dataframe
df = pd.read_csv("STEM_Salary_Data.csv")

# Define a function to convert years of experience to level
def convert_experience_to_level(years_of_experience):
    if years_of_experience < 2:
        return 'entry-level'
    elif years_of_experience >= 2 and years_of_experience <= 5:
        return 'mid-level'
    else:
        return 'senior'
    
# Add a new column level based on yearsofexperience
df['level'] = df['yearsofexperience'].apply(convert_experience_to_level)
```

```python
df.head()
```

Now, we will create a bar chart to see the number of people who are entry level, mid level, and are senior level.

### Using matplotlib

```python
# Import matplotlib
import matplotlib.pyplot as plt
```

```python
# Get the value counts of unique values in the 'level' column
value_counts = df['level'].value_counts()

# Plot a bar chart for number in each level
plt.bar(value_counts.index, value_counts)

# Add title and axis labels
plt.title('Distribution of Experience Levels')
plt.xlabel('Experience Level')
plt.ylabel('Number of People')

# Show the plot
plt.show()
```

### Use Pandas

```python
# Group the data by the level and find the number of people in each level
level_counts = df.groupby('level').size()
```

```python
# Create bar chart
level_counts.plot(kind='bar')

# Add title and axis labels
plt.title('Distribution of Experience Levels')
plt.xlabel('Experience Level')
plt.ylabel('Number of People')

# Show the plot
plt.show()
```

### Using Seaborn

```python
# import library
import seaborn as sns
```

```python
# Create a bar chart using Seaborn
sns.countplot(x = 'level', data=df)

# Add title and axis labels
plt.title('Distribution of Experience Levels')
plt.xlabel('Experience Level')
plt.ylabel('Number of People')

# Show the plot
plt.show()
```
