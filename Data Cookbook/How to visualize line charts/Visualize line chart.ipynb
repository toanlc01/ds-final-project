# How to visualize line charts

This dataset contains years of experience and its corresponding salary. We will learn to visualize it using the line charts

### Using matplotlib

```python
# Import the library
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# load the dataset into a pandas DataFrame
df = pd.read_csv('Salary_Data.csv')
```

```python
df.head()
```

```python
x_axis = df['YearsExperience']
y_axis = df['Salary']
```

```python
# create a line chart with years of experience on the x-axis and salary on the y-axis
plt.plot(x_axis, y_axis)

# add title and axis labels
plt.title('Salary vs Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')

# show the plot
plt.show()
```

### Using pandas

```python
# Create the line chart
df.plot(x='YearsExperience', y='Salary', kind='line')

# Add title and axis labels
plt.title('Salary by Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')

# Show the plot
plt.show()
```

### Using Seaborn

```python
# Import the library
import seaborn as sns

# Plot the line graph
sns.lineplot(data=df, x='YearsExperience', y='Salary')

# Add title and axis labels
plt.title('Salary by Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
```
