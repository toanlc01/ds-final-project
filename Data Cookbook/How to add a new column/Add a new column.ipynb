# Add a new column to an existing dataset

```python
# Import the libraries
import pandas as pd
```

```python
# Read the data into dataframe
df = pd.read_csv("STEM_Salary_Data.csv")
```

```python
df.head()
```

There's is a column `yearsofexperience` in the dataset. We will add a new column name level based on their years of experience. So if a person has less than 2 years of experience, we will categorize it as an entry level. If a person has from 2 to 5 years of experience, we will categorize it as a mid level. If a person has more than 5 years of experience, we will categorize it as a senior level

```python
# Define a function to convert years of experience to level
def convert_experience_to_level(years_of_experience):
    if years_of_experience < 2:
        return 'entry-level'
    elif years_of_experience >= 2 and years_of_experience <= 5:
        return 'mid-level'
    else:
        return 'senior'
```

```python
# Add a new column level based on yearsofexperience
df['level'] = df['yearsofexperience'].apply(convert_experience_to_level)
```
